# Importando a biblioteca

```python
import pandas as pd
```

Agora vamos abrir nosso arquivo csv em um dataframe para começarmos a trabalhar com os dados.

```python
df = pd.read_csv('https://storage.googleapis.com/bc26-bucket-aulas/ocorrencia.csv',sep=';',encoding = 'ISO-8859-1')

df
```

Nosso dataframe possui 22 colunas e 6769 linhas e contém dados de incidentes aeronáuticos, disponibilizados no site dados.gov. Constam nesta base de dados as ocorrências aeronáuticas notificadas ao CENIPA no anos de 2010 a 2022 que ocorreram em solo brasileiro.

Vamos iniciar um tratamento simples em nossos dados conhecer alguns métodos do Pandas. 

#### **head() e tail()**



Exibe os primeiros (head) ou últimos (tail) registros do dataframe. Dentro do parênteses você pode especificar a quantidade de linhas que deseja visualizar. 

```python
df.head(2)
```

```python
df.tail(2)
```

#### **shape**

Exibe quantas linhas e colunas existem no dataframe

```python
df.shape
```

#### **dtypes**

Exibe os tipos de dados das colunas do dataframe, por exemplo: float, int, object, datetime, etc.

```python
df.dtypes
```

#### **is_unique, unique(), nunique()**




São comandos que mostram elementos únicos do seu dataframe.

O **is_unique** tem retorno booleano e retorna True quando os elementos de uma coluna são únicos;


```python
df.codigo_ocorrencia.is_unique
```

O **unique()** retorno quais os elementos que são únicos na coluna

```python
df.ocorrencia_saida_pista.unique()
```

```python
sorted(pd.unique(df['ocorrencia_saida_pista']))
```

O **nunique()** conta os valores distintos das colunas. Isso mostra apenas o número de categorias diferentes presentes em uma coluna de um dataframe. 

```python
df['ocorrencia_saida_pista'].nunique()
```

#### **isnull()**

Verifica em qual linha da coluna exite um valor nulo, retornado True o False para cada uma delas.

```python
df['ocorrencia_saida_pista'].isnull()
```

#### **isna()**

Exibe a quantidade de valores ausentes em cada coluna

```python
#consulta valores ausentes
df.isna().sum()
```

#### **count()**

Mostra a quantidade de valores válidos em cada coluna.

```python
#consulta dados "válidos"
df.count()
```

#### **info()**
Mostra um resumo das informções do dataframe como: index, lista das colunas, quantidade de valores não nulos, tipo dos dados, uso de memória.

```python
df.info()
```

## Filtros

O **.loc** é utilizado para busca no seu dataframe

#### **loc**

```python
#busca uma fatia do dataframe da linha 80 a 100

df.loc[80:100]
```

```python
#podemos chamar uma linha pelo seu índice
df.loc[5]
```

```python
#ou com um array de índices
df.loc[[0,1,2]] 
```

```python
#passando a linha e a coluna que queremos buscar
df.loc[0,'ocorrencia_cidade']
```

```python
#sem especificar a linha, apenas a coluna
df.loc[:,'ocorrencia_cidade'] 
```

```python
#podemos usar para atribuir um novo valor para uma linha
df.loc[0,'ocorrencia_cidade'] = 'SÃO PAULO'
```

```python
#criando um filtro dos estados que estão preenchidos como nulos
filtrouf = df.ocorrencia_uf.isna()

#criando um filtro dos horários que estão preenchidos como nulos
filtrohorario = df.ocorrencia_hora.isna()
```

```python
#usando o loc com os filtros criados
df.loc[filtrouf]
```

```python
df.loc[filtrohorario]
```
